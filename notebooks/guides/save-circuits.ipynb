Usa la [serializzazione QPY](https://docs.quantum.ibm.com/api/qiskit/qpy) per salvare il tuo Circuit su file. I file QPY memorizzano l'intero oggetto Circuit di Qiskit e saranno compatibili con le versioni più recenti di Qiskit (ma non necessariamente con le versioni precedenti).

Per fare un esempio, la cella seguente crea un semplice Circuit quantistico.

````python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()
````

Per salvare questo file su disco, usa la funzione `qpy.dump`. Puoi anche salvare una lista di Circuit.

````python
from qiskit import qpy

with open("test.qpy", "wb") as file:
    qpy.dump(qc, file)
````

Questo Circuit è ora salvato nel file `test.qpy`. Se riavvii il kernel Python, puoi ricaricare il Circuit usando la funzione `qpy.load`. Nota che questa funzione restituisce sempre una lista di Circuit, anche se ne hai serializzato uno solo.

````python
with open("test.qpy", "rb") as handle:
    qc = qpy.load(handle)

qc[0].draw("mpl")
````
